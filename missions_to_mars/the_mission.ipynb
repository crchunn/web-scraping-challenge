```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
def init_browser():
    executable_path = {"executable_path": "../chromedriver"}
    return Browser("chrome", **executable_path, headless=True)
```

```python
def scrape():
    browser = init_browser()

    url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
    browser.visit(url)

    html = browser.html
    soup = BeautifulSoup(html, "html.parser")

    slide_elem = soup.select_one("ul.item_list li.slide")
    slide_title = slide_elem.find("div", class_='content_title').get_text()
    slide_body = slide_elem.find("div", class_='article_teaser_body').get_text()
   # listings["price"] = soup.find("span", class_="result-price").get_text()
   # listings["hood"] = soup.find("span", class_="result-hood").get_text()
    return slide_title, slide_body

scrape()
```

```python
def scrape_hemi():
    browser = init_browser()
    url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
    browser.visit(url)
    
    featured_image_url = browser.find_by_id("Cerberus Hemisphere Enhanced")
    featured_image_url.click()
    
    # Find and click the full image button

    # Find the more info button and click that
    browser.is_element_present_by_text('Sample', wait_time=1)
    more_info_elem = browser.links.find_by_partial_text('Sample')
    more_info_elem.click()
    # Parse the resulting html with soup
  
    html = browser.html
    soup = BeautifulSoup(html, "html.parser")

    cerb_fig = soup.select_one("div.downloads a img").get("src")
   
    return cerb_fig
scrape_hemi()
```

```python
def scrape_img():
    browser = init_browser()
    url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
    browser.visit(url)
    
    featured_image_url = browser.find_by_id("full_image")
    featured_image_url.click()
    
    # Find and click the full image button

    # Find the more info button and click that
    browser.is_element_present_by_text('more info', wait_time=1)
    more_info_elem = browser.links.find_by_partial_text('more info')
    more_info_elem.click()
    # Parse the resulting html with soup
  
    html = browser.html
    soup = BeautifulSoup(html, "html.parser")

    slide_fig = soup.select_one("figure.lede a img").get("src")
   
    return slide_fig
scrape_img()
```
